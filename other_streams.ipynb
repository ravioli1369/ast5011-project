# **Analyze Other Streams**

It is increasingly non trivial to obtain the stream stars for streams other than GD-1, so we use the preprocessed stream stars from the [`STREAMFINDER`](https://iopscience.iop.org/article/10.3847/1538-4357/ad382d#apjad382dt1) algorithm.

We also compute the density profile of the stream by counting the number of stars in bins along the phi1 axis. No need to perform background subtraction as we are using basically pure stream stars. Instead, we do a simple median subtraction to see where there are over or under-densities along the stream.

Refer `streams.py` for the helper functions.

## **Jhelum**

```python
jhelum = Streams.JHELUM
jhelum_data = obtain_stream_data(jhelum)
JHELUM_PHI_XLIM = (-10, 27)
JHELUM_PHI_YLIM = (-4, 6)
jhelum_density = obtain_stream_density(
    jhelum_data, PHI_XLIM=JHELUM_PHI_XLIM, binstep=0.5
)
plot_density(
    jhelum,
    jhelum_density,
    jhelum_data,
    PHI_XLIM=JHELUM_PHI_XLIM,
    PHI_YLIM=JHELUM_PHI_YLIM,
)
```
